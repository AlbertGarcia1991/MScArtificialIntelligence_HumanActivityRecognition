## 2. Data augmentation (feature engineering)

```python
# Rename DataFrame columns
phone_accel_columns = ['user', 'activity', 'P_LA_x', 'P_LA_y', 'P_LA_z']
phone_gyro_columns = ['user', 'activity', 'P_AV_x', 'P_AV_y', 'P_AV_z']
watch_accel_columns = ['user', 'activity', 'W_LA_x', 'W_LA_y', 'W_LA_z']
watch_gyro_columns = ['user', 'activity', 'W_AV_x', 'W_AV_y', 'W_AV_z']
phone_accel.columns = phone_accel_columns
phone_gyro.columns = phone_gyro_columns
watch_accel.columns = watch_accel_columns
watch_gyro.columns = watch_gyro_columns

# Generate new features integrating
try:
    phone_accel['P_LO_x'] = phone_accel['P_LA_x'].diff(periods=-1).fillna(0) / 0.05
    phone_accel['P_LO_y'] = phone_accel['P_LA_y'].diff(periods=-1).fillna(0) / 0.05
    phone_accel['P_LO_z'] = phone_accel['P_LA_z'].diff(periods=-1).fillna(0) / 0.05
except: pass

try:
    watch_accel['W_LO_x'] = watch_accel['W_LA_x'].diff(periods=-1).fillna(0) / 0.05
    watch_accel['W_LO_y'] = watch_accel['W_LA_y'].diff(periods=-1).fillna(0) / 0.05
    watch_accel['W_LO_z'] = watch_accel['W_LA_z'].diff(periods=-1).fillna(0) / 0.05
except: pass

try:
    phone_gyro['P_AA_x'] = phone_gyro['P_AV_x'].diff().fillna(0) / 0.05
    phone_gyro['P_AA_y'] = phone_gyro['P_AV_y'].diff().fillna(0) / 0.05
    phone_gyro['P_AA_z'] = phone_gyro['P_AV_z'].diff().fillna(0) / 0.05
except: pass

try:
    watch_gyro['W_AA_x'] = watch_gyro['W_AV_x'].diff().fillna(0) / 0.05
    watch_gyro['W_AA_y'] = watch_gyro['W_AV_y'].diff().fillna(0) / 0.05
    watch_gyro['W_AA_z'] = watch_gyro['W_AV_z'].diff().fillna(0) / 0.05
except: pass

try:
    phone_gyro['P_AO_x'] = phone_gyro['P_AA_x'].diff(periods=-1).fillna(0) / 0.05
    phone_gyro['P_AO_y'] = phone_gyro['P_AA_y'].diff(periods=-1).fillna(0) / 0.05
    phone_gyro['P_AO_z'] = phone_gyro['P_AA_z'].diff(periods=-1).fillna(0) / 0.05
except: pass

try:
    watch_gyro['W_AO_x'] = watch_gyro['W_AA_x'].diff(periods=-1).fillna(0) / 0.05
    watch_gyro['W_AO_y'] = watch_gyro['W_AA_y'].diff(periods=-1).fillna(0) / 0.05
    watch_gyro['W_AO_z'] = watch_gyro['W_AA_z'].diff(periods=-1).fillna(0) / 0.05
except: pass

# Convert all 4 original datasets into a dict of DataFrames indexed by pair subject-activity
df_dict = {}
for i in range(51):
    prov_dict = {}
    for j in range(18):
        user_prov = phone_accel[phone_accel['user']==i]
        pa_activity_prov = user_prov[user_prov['activity']==j]
        pa_activity_prov = pa_activity_prov.reset_index()
        
        user_prov = phone_gyro[phone_gyro['user']==i]
        pg_activity_prov = user_prov[user_prov['activity']==j].drop(columns=['user', 'activity'])
        pg_activity_prov = pg_activity_prov.reset_index()
        
        user_prov = watch_accel[watch_accel['user']==i]
        wa_activity_prov = user_prov[user_prov['activity']==j].drop(columns=['user', 'activity'])
        wa_activity_prov = wa_activity_prov.reset_index()
        
        user_prov = watch_gyro[watch_gyro['user']==i]
        wg_activity_prov = user_prov[user_prov['activity']==j].drop(columns=['user', 'activity'])
        wg_activity_prov = wg_activity_prov.reset_index()
        
        prov = pd.concat([pa_activity_prov, pg_activity_prov, wa_activity_prov, wg_activity_prov], axis=1, sort=False).dropna()
        
        prov_dict[j] = prov
    df_dict[i] = prov_dict
for i in range(51):
    for j in range(18):
        df_dict[i][j] = df_dict[i][j][df_dict[i][j]['user']==i].reset_index()
for i in range(51):
    for j in range(18):
        try:
            ROWS_TO_REMOVE = list(range(500)) + list(range(3100, len(df_dict[i][j]),1))
            df_dict[i][j] = df_dict[i][j].drop(ROWS_TO_REMOVE).reset_index()
        except: pass
for i in range(51):
    for j in range(18):
        try:
            df_dict[i][j] = df_dict[i][j].drop(columns=['index', 'level_0', 'user', 'activity'])
        except: pass
for i in range(51):
    for j in range(18):
        for column in ['P_LA_x', 'P_LA_y', 'P_LA_z', 'W_LA_x', 'W_LA_y', 'W_LA_z']:
            serie = []
            serie.append(0)
            for k in range(len(df_dict[i][j][column])):
                acc_value = serie[-1] + df_dict[i][j][column][k] * 0.05
                serie.append(acc_value)
            if column == 'P_LA_x':
                df_dict[i][j]['P_LV_x'] = pd.Series(serie[:-1])

            elif column == 'P_LA_y':
                df_dict[i][j]['P_LV_y'] = pd.Series(serie[:-1])

            elif column == 'P_LA_z':
                df_dict[i][j]['P_LV_z'] = pd.Series(serie[:-1])

            elif column == 'W_LA_x':
                df_dict[i][j]['W_LV_x'] = pd.Series(serie[:-1])

            elif column == 'W_LA_y':
                df_dict[i][j]['W_LV_y'] = pd.Series(serie[:-1])

            elif column == 'W_LA_z':
                df_dict[i][j]['W_LV_z'] = pd.Series(serie[:-1])
                
# Rename new features columns                
final_column_order = ['P_LO_x', 'P_LO_y', 'P_LO_z', 'P_LA_x', 'P_LA_y', 'P_LA_z', 'P_LV_x', 'P_LV_y', 'P_LV_z',
                      'P_AO_x', 'P_AO_y', 'P_AO_z', 'P_AA_x', 'P_AA_y', 'P_AA_z', 'P_AV_x', 'P_AV_y', 'P_AV_z',
                      'W_LO_x', 'W_LO_y', 'W_LO_z', 'W_LA_x', 'W_LA_y', 'W_LA_z', 'W_LV_x', 'W_LV_y', 'W_LV_z',
                      'W_AO_x', 'W_AO_y', 'W_AO_z', 'W_AA_x', 'W_AA_y', 'W_AA_z', 'W_AV_x', 'W_AV_y', 'W_AV_z']
for i in range(51):
    for j in range(18):
        try:
            df_dict[i][j] = df_dict[i][j][final_column_order]
        except: pass
```

## 3. Data normalization

```python
# Obtain the global max and min for aeach feature inspecting all users and all activities
maxmin_dict = {}
for column in final_column_order:
    maxmin_dict[column] = 0
flag = True
for i in range(51):
    for j in range(18):
        for column in final_column_order:
            if flag:
                max_value = df_dict[i][j][column].max()
                min_value = df_dict[i][j][column].min()
                maxmin_dict[column] = np.array((max_value, min_value))

            else:
                max_value = df_dict[i][j][column].max()
                min_value = df_dict[i][j][column].min()
                if max_value > maxmin_dict[column][0]:
                    maxmin_dict[column][0] = max_value
                if min_value < maxmin_dict[column][1]:
                    maxmin_dict[column][1] = min_value          
    if flag: flag=False

# Apply the normalization between -1 and +1
for i in range(51):
    for j in range(18):
        for column in final_column_order:
            try:
                df_dict[i][j][column] = 2*((df_dict[i][j][column] - maxmin_dict[column][1]) / (maxmin_dict[column][0] - maxmin_dict[column][1])) - 1
            except:
                pass
```

## 4. Data preparation
